A script for updating the python jobs list.  
This time, the data was scraped using Beautiful soup.


```python
#import the necessary libraries
from  bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
from html.parser import HTMLParser
import timeit
import csv
from datetime import datetime as dt
```

```python
#opening the last scrapped csv file stored on my github
def last_data_scrapped():
  file = requests.get('https://raw.githubusercontent.com/enyeneraph/Python-Jobs-Analysis/main/python-jobs.csv').text
  data = list(csv.reader(file.splitlines()))
  last_row = data[1]
  return last_row
```

```python
last_row = last_data_scrapped()
```

```python
#function for opening url and parsing html
def open_url(url):
  try:
    text = requests.get(url).text
    soup = BeautifulSoup(text, 'html.parser')
  except ConnectionError:
    print('Error while connecting')
  return soup
```

```python
def parse_text(soup):
  container = soup.find(class_= 'list-recent-jobs list-row-container menu')
  lists_ = container.find_all('li')
  loop_status = True
  last_date = dt.strptime(last_row[3], '%d %B %Y')
  for item in lists_:
    company_info = item.find(class_ = "listing-company-name")
    title = company_info.a.get_text()
    location = item.find(class_ = 'listing-location').get_text().strip()
    looking_for = item.find(class_ = 'listing-job-type').get_text().strip()
    date_posted = item.time.get_text().strip()
    category = item.find(class_ = 'listing-company-category').get_text().strip()
    


    #cleaning up 
    company_info = company_info.get_text().strip().splitlines()
    company_name = company_info[-1].strip()

    job_details = [title,company_name, location, date_posted, looking_for, category]

  # check if date corresponds with date from last scrapped data
  #converting the dates for easy check
    date_posted = dt.strptime(date_posted, '%d %B %Y')
    if ((date_posted == last_date) and (job_details == last_row)) or date_posted < last_date:
      loop_status = False
      break
  #   #if it breaks will it leave the function?
    
    jobs_list.append(job_details)
  return loop_status
  
```

```python
#function for extracting next page url
def next_page_url(loop_status, soup):
  if loop_status:
    container = soup.find(class_ = "pagination menu")
    next_url = 'https://www.python.org/jobs/' + container.find(class_ = 'next').a['href']
    return next_url
```

```python
jobs_list = []
```

```python
def scrapper(url):
  soup = open_url(url)
  loop_status = parse_text(soup)
  next_url = next_page_url(loop_status, soup)
  if next_url:
    scrapper(next_url)
  print(loop_status)
  return jobs_list
```

```python
#run only once
scrapper('https://www.python.org/jobs')
```

```python
#converting the lists of jobs and their descriptions  to a dataframe
import pandas as pd
columns = ('Job_Title', 'Company', 'Location', 'Date_Posted', 'Looking_for','Category')
df = pd.DataFrame(jobs_list, columns=columns)
```
